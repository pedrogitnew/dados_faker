```python
from faker import Faker
import random
import pandas as pd

# Inicializa o Faker em pt-BR
fake = Faker('pt_BR')

# Lista de produtos fictícios e seus preços
produtos = [
    {"nome": "Notebook", "preco": 3500.00},
    {"nome": "Mouse", "preco": 50.00},
    {"nome": "Teclado", "preco": 120.00},
    {"nome": "Monitor", "preco": 950.00},
    {"nome": "Cadeira Gamer", "preco": 1400.00},
    {"nome": "Headset", "preco": 300.00},
    {"nome": "Webcam", "preco": 200.00}
]

# Lista de estados brasileiros (UFs)
estados = [
    'AC', 'AL', 'AP', 'AM', 'BA', 'CE', 'DF', 'ES', 'GO', 'MA',
    'MT', 'MS', 'MG', 'PA', 'PB', 'PR', 'PE', 'PI', 'RJ', 'RN',
    'RS', 'RO', 'RR', 'SC', 'SP', 'SE', 'TO'
]

# Quantidade de registros
num_registros = 100

# Lista para armazenar os dados
dados_vendas = []

# Geração dos dados de venda como ID do pedido, Data, Nome do cliente, Sexo e etc
for i in range(1, num_registros + 1):
    produto = random.choice(produtos)
    quantidade = random.randint(1, 5)
    preco_unitario = produto["preco"]
    total = round(preco_unitario * quantidade, 2)

    sexo = random.choice(['Masculino', 'Feminino'])
    if sexo == 'Masculino':
        nome = fake.name_male()
    else:
        nome = fake.name_female()

    estado = random.choice(estados)
    
    dados_vendas.append({
        "ID Pedido": f"PED{i:04d}",
        "Data da Venda": fake.date_between(start_date='-6M', end_date='today').strftime('%d/%m/%Y'),
        "Cliente": nome,
        "Sexo": sexo,
        "CPF": fake.cpf(),
        "Estado": estado,
        "Produto": produto["nome"],
        "Quantidade": quantidade,
        "Preço Unitário (R$)": round(preco_unitario, 2),
        "Total da Venda (R$)": total
    })

# Converte para DataFrame
df = pd.DataFrame(dados_vendas)

# Exibe os 5 primeiros
df.head()

# (Opcional) Salvar em Excel
# df.to_excel('dados_vendas_com_estado_e_sexo.xlsx')
```
